```python
# import
import pandas as pd
import numpy as np
import json
from pandas.io.json import json_normalize
```

```python
#json 파일 불러오고 dataframe 형식으로

with open("./sample.json", "r", encoding="utf-8") as f:
    data = json.load(f)

sample = json_normalize(data)
sample
```

```python
df = pd.read_csv("2018_SEL_FILLED.csv") ## file reading
df = df.dropna(axis=0)
df.head()
df.info()
```

```python
df = df.astype({'HE_DM' : 'int'})
x = df.drop(columns=["HE_DM"])
y = df.HE_DM
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
def drop(train, test, bye = ['patientID']):
    for i in [train, test]:
        for z in bye:
            del i[z]
    return train, test

x_train_id = x_train.patientID
x_test_id = x_test.patientID
x_train, x_test = drop(x_train, x_test)
```

```python
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import accuracy_score

# RandomForest
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators = 100, criterion = 'entropy', random_state = 0, max_features = 'auto', max_depth = 10)
rf.fit(x_train, y_train)

pred_rf = rf.predict(x_test)
print(accuracy_score(y_test, pred_rf))
plot_confusion_matrix(rf, x_test, y_test)
```

```python
pred_rf = rf.predict(sample.drop(columns=["patientID"]))
pred_rf = pd.DataFrame(pred_rf, columns=['HE_DM'])
```

```python
pred_rf = pd.concat((sample.patientID, pred_rf), axis =1)
pred_rf
```

```python
pred_rf = pred_rf.to_json('pred_rf.json', orient='records')
```

```python
## Logistic Regression
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(random_state=0).fit(x_train, y_train)
pred_lr = lr.predict(x_test)
print(accuracy_score(y_test, pred_lr))
plot_confusion_matrix(lr, x_test, y_test)
```

```python
pred_lr = lr.predict(sample.drop(columns=["patientID"]))
pred_lr = pd.DataFrame(pred_lr, columns=['HE_DM'])

pred_lr = pd.concat((sample.patientID, pred_lr), axis =1)
pred_lr
```

```python
pred_lr = pred_lr.to_json('pred_lr.json', orient='records')
```
